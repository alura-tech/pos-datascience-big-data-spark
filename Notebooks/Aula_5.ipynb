###**Identificação de dados nulos**

```python
from pyspark.sql import functions as f
```

```python
animes.select([f.count(f.when(f.isnull(c),1)).alias(c) for c in animes.columns]).show()
```

```python
animes = animes.na.drop()
```

```python
animes.select([f.count(f.when(f.isnull(c),1)).alias(c) for c in animes.columns]).show()
```

###**Seleções e consultas com select()**

###**Ordenando os dados**

```python
animes\
  .groupBy('Genres')\
  .count()\
  .orderBy('count',ascending=False)\
  .show(10, False)
```

```python
animes.orderBy('Score',ascending=True).show(10,False)
```

###**Filtrando os dados com where()**

```python
animes.where('score > 8.80').toPandas().plot.barh(x='Name',y='Score', figsize= (19,12))
```

```python
animes.where('score > 8.80').to_pandas_on_spark().plot.barh(y='Name',x='Score')
```

```python
animes\
  .where('Name == "Kimetsu no Yaiba"')\
  .show(5, False)
```

###**Filtrando os dados com filter()**

```python
animes\
  .select('Name', 'Score')\
  .filter(animes.Name.startswith('Kimetsu'))\
  .show(5, False)
```

###**Utilizando like()**

```python
animes\
  .select('Name', 'Score', 'Genres')\
  .filter((animes['Name']).like('%Season'))\
  .orderBy('Name', ascending=True)\
  .show(15, False)
```

```python
animes\
  .select('Name', 'Score', 'Genres')\
  .filter((animes['Name']).like('%Season%'))\
  .orderBy('Name', ascending=True)\
  .show(20, False)
```

```python
animes\
  .select('Name', 'Score', 'Genres')\
  .filter((animes['Name']).like('Season%'))\
  .orderBy('Name', ascending=True)\
  .show(15, False)
```

image.png

```python
categorias = spark.createDataFrame(
    [
     ('1','Hunter x Hunter','Ação'),
     ('2','Dragon Ball', 'Ação'),
     ('3','Naruto','Ação'),
     ('4','Death Note','Suspense'),
     ('5', 'Steins Gate','Ficção')],
    ['id','nome','categoria'])
```

```python
notas = spark.createDataFrame(
    [
     ('Hunter x Hunter',9.15),
     ('Dragon Ball',9.75),
     ('Naruto',10.00),
     ('Death Note',10.00)],
    ['nome','nota'])
```

```python
categorias.join(notas,'nome',how='left')\
  .sort('id')\
  .show()
```

```python
categorias.join(notas,'nome',how='right')\
  .sort('id')\
  .show()
```

#**Aula 5 - Análise exploratória dos Datasets**

###**Quais são os gêneros mais vistos?**

```python
df\
  .groupBy('Genres')\
  .count()\
  .orderBy('count',ascending=False)\
  .show(10, False)
```

###**Quais são os animes mais vistos?**

```python
df\
  .groupBy('Name')\
  .count()\
  .orderBy('count',ascending=False)\
  .show(10, False)
```

###**Quais são os animes com maiores médias de avaliação?**

```python
anime_avg_rating = df.groupBy('Name')\
    .agg(f.count(f.col('user_id')).alias('user_cnt'),
         f.mean(f.col('rating')).alias('mean_rating')).orderBy('mean_rating', ascending=False)
```

```python
anime_avg_rating.show(10, False)
```

```python
anime_avg_rating = df.groupBy('anime_id', 'Name', 'Genres')\
    .agg(f.count(f.col('user_id')).alias('user_cnt'),
         f.mean(f.col('rating')).alias('mean_rating')).orderBy('mean_rating', ascending=False)
```

```python
anime_avg_rating\
    .filter(animes.Genres.contains('Demons'))\
    .orderBy('mean_rating', ascending=False)\
    .show(truncate=False)
```
