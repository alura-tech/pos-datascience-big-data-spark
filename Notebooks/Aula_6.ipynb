#**Aula 6 - Recomendações com o Algoritmo ALS**

###**Importações**

```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.recommendation import ALS 
```

###**Definindo quais colunas serão usadas:**

```python
data = ratings.select(ratings['anime_id'], ratings['rating'], ratings['user_id'])
data.show()
```

###**Separando os dados em treino e teste:**

```python
(training, test) = data.randomSplit([.8, .2], seed=7589)
```

###**Parâmetros do modelo:**

```python
als = ALS(maxIter=5, rank=4, regParam=0.10, userCol='user_id', itemCol='anime_id',
          ratingCol='rating', coldStartStrategy='drop')
```

###**Fit do modelo de recomendação usando ALS nos dados de treinamento:**

```python
model = als.fit(training)
```

###**Avaliação do modelo calculando o RMSE nos dados de teste:**

```python
predictions = model.transform(test)
evaluator = RegressionEvaluator(metricName='rmse', labelCol='rating', predictionCol='prediction')
rmse = evaluator.evaluate(predictions)
print('RMSE = ' + str(rmse))
```

###**Predições nos dados de teste**:

```python
animes = animes.drop('Score')
```

```python
predictions = model.transform(test)
predictions.join(animes, 'anime_id', how='inner').show(20,False)
```

###**Recomendações**

```python
recommend_all_users = model.recommendForAllUsers(10)
recommend_all_users.show(20, False)
```

```python
USER_ID = 22
(recommend_all_users
  .filter(f'user_id == {USER_ID}')
  .withColumn('rec', f.explode('recommendations'))
  .select('user_id', 
          f.col('rec').anime_id.alias('anime_id'),
          f.col('rec').rating.alias('rating'),
          )
  .join(animes, 'anime_id')
  .orderBy('rating', ascending=False)
  .select('anime_id', 'Name', 'Genres')).show(truncate=False)
```
